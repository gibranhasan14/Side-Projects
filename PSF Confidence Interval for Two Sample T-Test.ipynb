# Confidence Interval for two sample T-Test to compare two means

```python
diff = []

for _ in range(10000):
    bootsample = bookings_random.sample(152, replace=True)
    mean_coff = bootsample[bootsample['PSF_status'] == 'No PSF Funds Requested']['MonthlyAvgBookings'].mean()
    mean_nocoff = bootsample[bootsample['PSF_status'] == 'PSF Funds Requested']['MonthlyAvgBookings'].mean()
    diff.append(mean_coff - mean_nocoff)

np.percentile(diff, 2.5), np.percentile(diff, 97.5)
```

# Hypothesis Testing

```python
nocoff_means, coff_means, diffs = [], [], []

for _ in range(10000):
    bootsample = bookings_random.sample(206, replace=True)
    coff_mean = bootsample[bootsample['PSF_status'] == 'No PSF Funds Requested']['MonthlyAvgBookings'].mean()
    nocoff_mean = bootsample[bootsample['PSF_status'] == 'PSF Funds Requested']['MonthlyAvgBookings'].mean()
    # append the info 
    coff_means.append(coff_mean)
    nocoff_means.append(nocoff_mean)
    diffs.append(coff_mean - nocoff_mean) 
```

```python
np.std(nocoff_means) # the standard deviation of the sampling distribution for nocoff
```

```python
np.std(coff_means) # the standard deviation of the sampling distribution for coff
```

```python
np.std(diffs) # the standard deviation for the sampling distribution for difference in means
```

```python
nocoff_mean
```

```python
coff_mean
```

```python
plt.hist(nocoff_means, alpha = 0.5);
plt.hist(coff_means, alpha = 0.5); # They look pretty normal to me!
```

```python
plt.hist(diffs, alpha = 0.5); # again normal - this is by the central limit theorem
```

```python
null_vals = np.random.normal(nocoff_mean, np.std(diffs), 10000) # Here are 10000 draws from the sampling distribution under the null
```

```python
null_vals
```

```python
plt.hist(null_vals); #Here is the sampling distribution of the difference under the null
```

# Two Sample T-Test

```python
import scipy
```

```python
scipy.stats.ttest_ind(nocoff_means, coff_means, equal_var=False)
```

```python
scipy.stats.ttest_ind(coff_means, nocoff_means, equal_var=False)
```

```python
##big t score means the means are extremely different from each other and p value of 0 means there is 0% chance this
##result occcurred by chance. 
```

```python
(null_vals > diffs).mean()
#favors null hypothesis that coff_mean 'No PSF Funds Requested' is > than nocoff_mean ' PSF Funds Requested'
```
